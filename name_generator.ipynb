# This is a Date generator

```
import datetime
import random
start_date = datetime.date(1960, 1, 1)
end_date = datetime.date(2001, 12, 1)
name_list = []

time_between_dates = end_date - start_date
days_between_dates = time_between_dates.days

for i in range(1000):
  random_number_of_days = random.randrange(days_between_dates)
  random_date = start_date + datetime.timedelta(days=random_number_of_days)
  random_date = random_date.strftime("%d-%b-%Y")
  name_list.append(random_date)
  
print(name_list)
df = pd.DataFrame({'birth_dates': name_list})
df.to_csv('birth_dates.csv')
```

# Password Generator

```
import random

print('''
Password Generator
==================
''')

chars = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ!.0123456789'

number = input('number of passwords?\n')
number = int(number)

length = input('password length?\n')
length = int(length)

pass_list = []

print('\nhere are your passwords:')

for pwd in range(number):
  password = ''
  for c in range(length):
    password += random.choice(chars)
  pass_list.append(password)
  
print(pass_list)
pass_df = pd.DataFrame({'passwords_generated': pass_list})
pass_df.to_csv('passwords_generated.csv')
```
